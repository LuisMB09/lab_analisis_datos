```python
df = pd.read_csv('data/new_users_forecast.csv', parse_dates=['month'])
df.set_index('month', inplace=True)
df.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df['new_users'], marker='o')
plt.title('Nuevos usuarios mensuales')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

```python
model = auto_timeseries(
    score_type='rmse',       # Métrica de evaluación
    time_interval='M',       # Intervalo mensual
    non_seasonal_pdq=None,   # Para modelos SARIMAX
    seasonality=True,       # Deshabilitar búsqueda de estacionalidad (se puede cambiar)
    model_type='best',       # Seleccionar el mejor modelo
    verbose=2                # Nivel de verbosidad
)
```

```python
train_size = int(0.8 * len(df))
train_df = df[:train_size]
test_df = df[train_size:]

model.fit(
    traindata=train_df,
    ts_column=train_df.index.name,
    target='new_users'
)
```

```python
forecast = model.predict(testdata=test_df)
forecast.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(train_df.index, train_df['new_users'], label='Entrenamiento')
plt.plot(test_df.index, test_df['new_users'], label='Datos Reales')
plt.plot(test_df.index, forecast['yhat'], label='Predicción', linestyle='--')
plt.title('Predicción de Usuarios Nuevos')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
rmse = mean_squared_error(test_df['new_users'], forecast['yhat'], squared=False)
print(f'RMSE en el conjunto de prueba: {rmse:.2f}')
```

## ARIMA

```python
train_size_ar = int(0.9 * len(df))
train_df_ar = df[:train_size_ar]
test_df_ar = df[train_size_ar:]

model_arima = pm.auto_arima(train_df_ar, 
                      seasonal=True,    
                      m=12,           
                      stepwise=True, 
                      trace=True)  
```

```python
forecast_arima = model_arima.predict(n_periods=len(test_df_ar))
forecast = pd.DataFrame()
forecast['ARIMA'] = forecast_arima
```

```python
forecast
```

```python
plt.figure(figsize=(12,6))

plt.plot(train_df_ar.index, train_df_ar['new_users'], label='Entrenamiento')
plt.plot(test_df_ar.index, test_df_ar['new_users'], label='Datos Reales')

plt.plot(forecast.index, forecast['ARIMA'], linestyle='--', label='ARIMA')
plt.title('Ususarios nuevos 3 años', fontsize=18)
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.legend(loc='best')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

## Definición de pruebas

**Prueba ARPU:**

- $H_0: \Delta ARPU \leq 300$

- $H_i: \Delta ARPU > 300$


**Prueba Conversion:**

$H_0: \Delta Conversión \leq 1\% $

$H_i: \Delta Conversión > 1\% $

**Calcular tamaño de muestra**:

$
n = \frac{2 \times (Z_{\alpha/2} + Z_{\beta})^2 \times \sigma^2}{\Delta^2}
$

```python
# Tamaño de muestra para el arpu
var = np.var(df_arpu['arpu'])
n = (2*(1.96+0.84)**2*var)/(300**2)
print(f'Tamaño de muestra para la prueba: {n}')
```

**Calcular tamaño de muestra**:

$
n = \frac{2 \times (Z_{\alpha/2} + Z_{\beta})^2 \times p(1-p)}{(p_1 - p_2)^2}
$

```python
p = df_conversion['converted'].mean()

n_p = (2*(1.96+0.84)**2*p*(1-p))/(0.01)**2
print(f'Tamaño de muestra para la prueba: {n_p}')
```

```python
df_ab_arpu = pd.read_csv('data/correcaminos_resultado_arpu_ab_test.csv')
df_ab_conv = pd.read_csv('data/correcaminos_resultado_conversion_ab_test.csv')
```

```python
df_arpu_A = df_ab_arpu.query('group == "A"')
df_arpu_B = df_ab_arpu.query('group == "B"')
df_arpu_A
```

```python
# Objectivo: tener diferencia de 20 dolares en el LTV
objective_difference = 300

# Promedio y desviación estandar por grupo
mean_a = np.mean(df_arpu_A['arpu'])
mean_b = np.mean(df_arpu_B['arpu'])

std_a = np.std(df_arpu_A['arpu'], ddof=1)
std_b = np.std(df_arpu_B['arpu'], ddof=1)

n_a = 980#len(df_arpu_A['arpu'])
n_b = 560# len(df_arpu_B['arpu'])

# Realizar un t-test de una cola (mayor o igual que el objetivo)
# Null hypothesis H0: (mean_b - mean_a) <= objective_difference
# Alternative hypothesis H1: (mean_b - mean_a) > objective_difference

# Calculate the t-statistic and p-value
pooled_std = np.sqrt(((std_a**2 / n_a) + (std_b**2 / n_b)))
t_stat = (mean_b - mean_a - objective_difference) / pooled_std
p_value = 1 - stats.t.cdf(t_stat, df=min(n_a, n_b) - 1)

# Output the results
print(f"Mean LTV of Group A: {mean_a:.2f}")
print(f"Mean LTV of Group B: {mean_b:.2f}")
print(f"Objective Difference: {objective_difference}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value <= 0.05:
    print('Se rechaza la hipótesis nula (H0)')
else:
    print('No se rechaza la hipótesis nula (H0)')
```

- $H_0: \Delta ARPU \leq 300$

- $H_i: \Delta ARPU > 300$

El aumento en el arpu es menor a $300, no se cumple el objetivo.

```python
# Objectivo: tener diferencia de 20 dolares en el LTV
objective_difference = 228

# Promedio y desviación estandar por grupo
mean_a = np.mean(df_arpu_A['arpu'])
mean_b = np.mean(df_arpu_B['arpu'])

std_a = np.std(df_arpu_A['arpu'], ddof=1)
std_b = np.std(df_arpu_B['arpu'], ddof=1)

n_a = 980#len(df_arpu_A['arpu'])
n_b = 560# len(df_arpu_B['arpu'])

# Realizar un t-test de una cola (mayor o igual que el objetivo)
# Null hypothesis H0: (mean_b - mean_a) <= objective_difference
# Alternative hypothesis H1: (mean_b - mean_a) > objective_difference

# Calculate the t-statistic and p-value
pooled_std = np.sqrt(((std_a**2 / n_a) + (std_b**2 / n_b)))
t_stat = (mean_b - mean_a - objective_difference) / pooled_std
p_value = 1 - stats.t.cdf(t_stat, df=min(n_a, n_b) - 1)

# Output the results
print(f"Mean LTV of Group A: {mean_a:.2f}")
print(f"Mean LTV of Group B: {mean_b:.2f}")
print(f"Objective Difference: {objective_difference}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value <= 0.05:
    print('Se rechaza la hipótesis nula (H0)')
else:
    print('No se rechaza la hipótesis nula (H0)')
```

- $H_0: \Delta ARPU \leq 228$

- $H_i: \Delta ARPU > 228$

El incremento en el ARPU es mayor a $228 al modificar los precios.

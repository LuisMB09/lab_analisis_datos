```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
```

```python
data = pd.read_csv("optitech_solutions_data.csv")
data['Date'] = pd.to_datetime(data['Date'])
fecha_filtro = '2022-01-31'
df = data.query('Date >= @fecha_filtro')
df.head()
```

```python
# Calculo cac
df['cac']= df['Marketing_Expenditure'] / df['New_Customers']
plt.plot(df.Date,df.cac, c = 'navy')
plt.grid()
```

```python
# calculo csat
df['csat'] = df['High_Customer_Score_Respondents']/df['Total_Customer_Respondents']
plt.plot(df.Date, df.csat, c = 'navy')
plt.grid()
```

```python
# Calculo eNPS
df['%_Employee_Promoter'] = df['Promoters_Employee_Score_Respondents']/df['Total_Employee_Respondents']
df['%_Employee_Detractor'] = df['Detractors_Employee_Score_Respondents']/df['Total_Employee_Respondents']
df['eNPS'] = df['%_Employee_Promoter'] - df['%_Employee_Detractor']
plt.plot(df.Date, df.eNPS, c='navy')
plt.grid()
```

```python
X = df[['Marketing_Expenditure','Web_Traffic']]  
y = df['Monthly_Revenue']  
```

```python
# Agregar la constante (intercepto) a las variables independientes
X = sm.add_constant(X)

# Ajustar el modelo usando OLS
model = sm.OLS(y, X)       
results = model.fit()      

# Mostrar el resumen del modelo
print(results.summary())
```
